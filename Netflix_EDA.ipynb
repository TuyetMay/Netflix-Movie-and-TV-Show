## Merge

```python
to_drop = netflix_titles[netflix_titles['show_id'].isna()].index

netflix_titles.drop(to_drop, inplace=True, axis=0)
netflix_titles['show_id'] = netflix_titles['show_id'].astype(int)
```

```python
show_id = netflix_titles['show_id'].astype(int).tolist()
```

```python
mapping_table = {
    'director': 'netflix_titles_directors',
    'cast': 'netflix_titles_cast',
    'country': 'netflix_titles_countries',
    'listed_in': 'netflix_titles_category'
}
for key in mapping_table:
    sheet = pd.read_excel(r"D:\netflix_titles.xlsx", sheet_name = mapping_table[key])
    tmp = []
    for id in show_id:
        query_result = sheet[sheet['show_id'] == id][key].tolist()
        if len(query_result) == 0:
            tmp.append(np.nan)
        else:
            tmp.append(query_result)
    netflix_titles[key] = tmp
```
